```python
import pandas as pd
import numpy as np
import scipy.io
import matplotlib.pyplot as plt
```

```python
#import data
df = pd.read_csv('T1D_association.csv', sep=',')
```

```python
#remove regex from variants column
df['Variant and risk allele'] = df['Variant and risk allele'].replace('<b>','',regex=True)
df['Variant and risk allele'] = df['Variant and risk allele'].replace('</b>','',regex=True)
```

```python
##splitting variant and risk allele into seperate columns 
df[['Variant and risk allele','Risk Allele']] = df['Variant and risk allele'].str.split('-',expand=True)
```

```python
#renaming variant... column to rsid 
df.rename({'Variant and risk allele': 'rsID'}, axis =1, inplace=True)
display(df)
```

```python
#filtering to isolate chr6
df[['Chromosome','Locality']] = df.Location.str.split(":", expand=True)
chromosome6_df = df.loc[df['Chromosome'] == '6']
```

```python
#DISPLAY DATAFRAME
display(chromosome6_df)
```

```python
chromosome6_df.to_csv('chr6_df_cleaned.csv', sep='\t')
```
